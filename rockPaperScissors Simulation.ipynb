```python
from os import system
import random
```

```python
nbr = 50000
```

```python
# Global Variables
null = 0
compWins = 0
playerWins = 0
```

```python
# Display Function
def PlayerChoise():
    print('Choose One :\n\n\t1) Rock\n\t2) Paper\n\t3) Scissors\n')
    x = int(input('Your Choose : 1 or 2 or 3 : '))
    return x
```

```python
# Random Conputer Choise Function
def conputerChoise():
    return random.randint(1, 3)
```

```python
p = PlayerChoise()
for i in range(nbr):
    c = conputerChoise()
    
    if c == p:
        null += 1
    elif (c == 1 and p == 3) or (c == 2 and p == 3) or (c ==3 and p == 2):
        compWins += 1
    else:
        playerWins += 1
```

```python
print('\nComputer :', compWins, ' Player :', playerWins, ' Null :', null)
```

```python
print('\nComputer :', (compWins / nbr) * 100, '% Player :', (playerWins / nbr) * 100, '% Null :', (null / nbr) * 100, '%')
```
